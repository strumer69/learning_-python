### using lambda in filters & maps

```python
# Program to filter out only the even items from a list

my_list = [1, 5, 4, 6, 8, 11, 3, 12]
```

```python
even_list = list(filter(lambda x:(x%2==0), my_list))
```

```python
even_list
```

# Exception Handling

```python
def divided_by(lis, a):
    return [i/a for i in lis]
        
```

```python
divided_by([1,2,3,4], 2)
```

```python
divided_by([1,2,3,4], 0)
```

```python
a = divided_by([1,2,3,4], 3)
print('Next line')
```

```python
a = divided_by([1,2,3,4], 0)
print('Next line')
```

```python
try:
    a = divided_by([1,2,3,4], 0)
except:
    print('An Error occured')
print('Next line')
```

```python
def divided_by_2(lis, a):
    try:
        return [i/a for i in lis]
    except:
        print('An error occured')
```

```python
a = divided_by_2([1,2,3,4], 0)
print('Next line')
```

```python
def divided_by_2(lis, a):
    try:
        return [i/a for i in lis]
    except ZeroDivisionError as e:
        
        print('An error occured: {}'.format(e))
    except: 
        print('An error occured!')
```

```python
a = divided_by_2([1,2,3,4], 0)
print('Next line')
```

### Finally

```python
def divided_by_2(lis, a):
    try:
        return [i/a for i in lis]
    except ZeroDivisionError as e:
        print('An error occured: {}'.format(e))
    finally: 
        print('All done')
        
```

```python
a = divided_by_2([1,2,3,4], 2)
print('Next line')
```

```python
a = divided_by_2([1,2,3,4], 0)
print('Next line')
```

```python
def divided_by_2(lis, a):
    try:
        return [i/a for i in lis]
    except ZeroDivisionError as e:
        print('An error occured: {}'.format(e.))
    print('All done')
```

```python
a = divided_by_2([1,2,3,4], 0)
print('Next line')
```

```python
a = divided_by_2([1,2,3,4], 2)
print('Next line')
```

```python
def divided_by_2(lis, a):
    try:
        return [i/a for i in lis]
    except ZeroDivisionError as e:
        print('An error occured: {}'.format(e.with_traceback))
    
```

```python
a = divided_by_2([1,2,3,4], 0)
print('Next line')
```

```python
def divided_by_2(lis, a):
    try:
        return [i/a for i in lis]
    except Exception as e:
        print('An error occured: {}'.format(e.with_traceback))
```

```python
a = divided_by_2([1,2,3,4], 0)
print('Next line')
```

### initializer or constructor

```python
class Employee:
    def __init__(self , first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        print('Employee {} {} was genrated'.format(self.first, self.last))
```

### adding some methods to class
#### Note: Every method defined in the class gets the instance of the class as an argument. So always use self!

```python
class Employee:
    def __init__(self , first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        print('Employee {} {} was genrated'.format(self.first, self.last))
    def get_email(self):
        return self.email
```

```python
emp1 = Employee('Mohammad', 'Amini', 50000)
emp2 = Employee('James', 'Bond', 45000)
```

```python
emp1.get_email()
```

```python
Employee.get_email(emp1)
```

```python
Employee.get_email(emp2)
```

```python
class Employee:
    def __init__(self , first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        print('Employee {} {} was genrated'.format(self.first, self.last))
    def get_email(self):
        return self.email
    def raise_pay(self):
        self.pay = int(self.pay* 1.04)
```

```python
emp1 = Employee('Mohammad', 'Amini', 50000)
```

```python
print(emp1.pay)
emp1.raise_pay()
print(emp1.pay)
```
